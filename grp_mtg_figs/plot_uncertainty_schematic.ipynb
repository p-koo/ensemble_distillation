```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.linspace(-8, 8, 100)
y = 3 * np.sin(x/2) + 2 * np.cos(x)

# Add some noise to create data points
np.random.seed(42)
x_data = np.linspace(-8, 8, 50)
y_data = 5 * np.sin(x_data/2) + 2 * np.cos(x_data) + np.random.normal(0, 1, 50)

# Create the plot
plt.figure(figsize=(4,3))

# Plot data points
plt.scatter(x_data, y_data, color='gray', alpha=0.7, label='Data points', s=4)

# Plot aleatoric uncertainty
plt.fill_between(x[x > 0], y[x > 0] - 2, y[x > 0] + 4, alpha=0.3, color='red', label='Aleatoric Uncertainty')

# Plot epistemic uncertainty
plt.fill_between(x[x < -4], y[x < -4] - 3, y[x < -4] + 3, alpha=0.3, color='blue', label='Epistemic Uncertainty')

# Set labels and title
plt.xlabel('X', size=14)
plt.ylabel('y', size=14)
# plt.title('Aleatoric and Epistemic Uncertainties')
plt.legend(frameon=False, loc='lower right')

# Set axis limits
plt.xlim(-8, 8)
plt.ylim(-15, 15)

# Remove ticks
plt.tick_params(axis='both', which='both', length=0,
                labelbottom=False, labelleft=False)

# Remove top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Show the plot
# plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.linspace(-8, 8, 100)
y = 5 * np.sin(x/2) + 2 * np.cos(x)

# Add some noise to create data points
np.random.seed(42)
x_data = np.linspace(-8, 8, 50)
y_data = 5 * np.sin(x_data/2) + 2 * np.cos(x_data) + np.random.normal(0, 1, 50)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot data points with larger size
point_size = 30  # You can adjust this value to make points larger or smaller
plt.scatter(x_data, y_data, color='gray', alpha=0.7, label='Data points', s=point_size)

# Plot aleatoric uncertainty
plt.fill_between(x[x > 0], y[x > 0] - 2, y[x > 0] + 2, alpha=0.3, color='red', label='Aleatoric Uncertainty')

# Plot epistemic uncertainty
plt.fill_between(x[x < -4], y[x < -4] - 3, y[x < -4] + 3, alpha=0.3, color='blue', label='Epistemic Uncertainty')

# Set labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Aleatoric and Epistemic Uncertainties')
plt.legend()

# Set axis limits
plt.xlim(-8, 8)
plt.ylim(-15, 15)

# Remove ticks and labels
plt.tick_params(axis='both', which='both', length=0, labelbottom=False, labelleft=False)

# Remove top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Show the plot
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
