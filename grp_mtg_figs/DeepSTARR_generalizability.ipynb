Plot performance of DeepSTARR models on sequences with OOD shift.

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

# MSE

```python
mse_df = pd.read_csv('../figures/DeepSTARR_ood_plotdata.csv')
```

```python
mse_df = mse_df[mse_df['Model2'].isin(['individual','distilled'])]
```

## Dev

```python
enhancer = 'Dev'
fig, ax = plt.subplots(1,figsize=(12,4))

# order = ['mutagenesis','evoaug','random']
order = ['small','intermediate','large']

# Dev
sns.boxplot(data=mse_df[mse_df['Enhancer']==enhancer], y='MSE',x='OOD shift', hue='Model2', ax=ax, fill=False, legend=False, order=order)
sns.swarmplot(data=mse_df[mse_df['Enhancer']==enhancer], y='MSE',x='OOD shift', hue='Model2', ax=ax, dodge=True, legend=True, order=order)
ax.set_xlabel('OOD shift')
ax.set_ylabel('MSE')
ax.set_title(f'STARR-seq: {enhancer}')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.legend(title='DeepSTARR', frameon=False)

fig.tight_layout()
fig.show()
```

## Hk

```python
enhancer = 'Hk'
fig, ax = plt.subplots(1,figsize=(12,4))

# order = ['mutagenesis','evoaug','random']
order = ['small','intermediate','large']

# Hk
sns.boxplot(data=mse_df[mse_df['Enhancer']==enhancer], y='MSE',x='OOD shift', hue='Model2', ax=ax, fill=False, legend=False, order=order)
sns.swarmplot(data=mse_df[mse_df['Enhancer']==enhancer], y='MSE',x='OOD shift', hue='Model2', ax=ax, dodge=True, legend=True, order=order)
ax.set_xlabel('OOD shift')
ax.set_ylabel('MSE')
ax.set_title(f'STARR-seq: {enhancer}')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.legend(title='DeepSTARR', frameon=False)

fig.tight_layout()
fig.show()
```

# Plot together

```python
fig, axs = plt.subplots(2,1, figsize=(8,6))

# order = ['mutagenesis','evoaug','random']
order = ['small','intermediate','large']

# Dev
sns.boxplot(data=mse_df[mse_df['Enhancer']==enhancer], y='MSE',x='OOD shift', hue='Model2', ax=axs[0], fill=False, legend=False, order=order)
sns.swarmplot(data=mse_df[mse_df['Enhancer']==enhancer], y='MSE',x='OOD shift', hue='Model2', ax=axs[0], dodge=True, legend=True, order=order)

for ax in axs:
    ax.set_xlabel('OOD shift')
    ax.set_ylabel('MSE')
    ax.set_title(f'STARR-seq: {enhancer}')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.legend(title='DeepSTARR', frameon=False)
```
